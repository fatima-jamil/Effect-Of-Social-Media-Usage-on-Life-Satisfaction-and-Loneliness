<center><font color="Violet" size="5"><b>Linear Regression Model To Predict Loneliness and Life Satisfaction</b></font>

```python
data=pd.read_csv("data_for_lr.csv")
```

```python
data.shape
```

#### Splitting Data

```python
train_input=np.array(data.Social_Media_Usage[0:90]).reshape(90,1)
train_output1=np.array(data.Lonelines_Score[0:90]).reshape(90,1)
train_output2=np.array(data.Life_Satisfaction[0:90]).reshape(90,1)

test_input=np.array(data.Social_Media_Usage[90:100]).reshape(9,1)
test_output1=np.array(data.Lonelines_Score[90:100]).reshape(9,1)
test_output2=np.array(data.Life_Satisfaction[90:100]).reshape(9,1)
```

```python
train_input.shape
```

```python
from sklearn.linear_model import LinearRegression
linear_regressor1=LinearRegression()
linear_regressor2=LinearRegression()
linear_regressor1.fit(train_input,train_output1)
linear_regressor2.fit(train_input,train_output2)
```

#### Parameters

```python
a=linear_regressor1.coef_
print(a)
```

```python
b=linear_regressor1.intercept_
print(b)
```

<font color="red"><b>Linear Regression equation for Loneliness Prediction = </font>a*(Social_Media_Usage_time) + b<br>


```python
c=linear_regressor2.coef_
print(c)
```

```python
d=linear_regressor2.intercept_
print(d)
```

<font color="red"><b>Linear Regression equation for Life Satisfaction Prediction = </font>c*(Social_Media_Usage_time) + d<br>


Predecting Test Input

```python
predicted_value1=linear_regressor1.predict(test_input)
predicted_value2=linear_regressor2.predict(test_input)
```

```python
from sklearn.metrics import mean_squared_error
cost1=mean_squared_error(test_output1,predicted_value1)
cost2=mean_squared_error(test_output2,predicted_value2)
```

```python
cost1
```

```python
cost2
```

#### Visualising The Model Performance

test_input=Social Media Usage<br>
test_output1=predicted_value1=Lonelines Score<br>
test_output2=predicted_value2=Life Satisfaction

```python
plt.plot(test_input,test_output1,'+',color="green")
plt.plot(test_input,predicted_value1,'*',color="red")
plt.title('Performance Testing')
plt.xlabel('input')
plt.xlabel('output')
plt.show()
```

```python
plt.plot(test_input,test_output2,'+',color="green")
plt.plot(test_input,predicted_value2,'*',color="red")
plt.title('Performance Testing')
plt.xlabel('input')
plt.xlabel('output')
plt.show()
```

```python
import statsmodels.api as sm
```

```python
x_train=sm.add_constant(train_input)
model1=sm.OLS(train_output1,x_train)
result1=model1.fit()
model2=sm.OLS(train_output2,x_train)
result2=model2.fit()
```
